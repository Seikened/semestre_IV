


# Crear una tabla fácil
Se crea una tabla demo con dos ejes de una manera muy sencilla


```python

import matplotlib.pyplot as plt
import numpy as np

rango = 3

x = np.linspace(-rango,rango,100)
y = np.linspace(-rango,rango,100)


fig,ax = plt.subplots(figsize=(10,5),layout='constrained')

ax.scatter(x,y,color="skyblue")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("Soy el titulo de tu gráfica")
plt.show()
```

## Para mejorar la tabla
Aquí mejoraremos todos los campos

```python

import matplotlib.pyplot as plt
import numpy as np

# Aplicamos un estilo visual bonito (puedes cambiarlo: 'seaborn', 'ggplot', 'fivethirtyeight', etc.)
plt.style.use('seaborn-v0_8-whitegrid')

rango = 3
x = np.linspace(-rango, rango, 100)
y = np.linspace(-rango, rango, 100)

fig, ax = plt.subplots(figsize=(10, 6), layout='constrained')  # Tamaño mejorado

# Scatter mejorado con transparencia, borde y tamaño
ax.scatter(x, y, 
           color="skyblue", 
           edgecolor='steelblue', 
           s=80,  # tamaño de punto
           alpha=0.7,  # transparencia
           label="Puntos generados")

# Límites para que no esté pegado a los bordes
ax.set_xlim(-rango - 1, rango + 1)
ax.set_ylim(-rango - 1, rango + 1)

# Etiquetas mejoradas
ax.set_xlabel("Eje X", fontsize=14)
ax.set_ylabel("Eje Y", fontsize=14)
ax.set_title("🌌 Mi hermosa gráfica Scatter", fontsize=16, weight='bold')

# Activamos la cuadrícula y eje centrado si quieres
ax.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)

# Opción extra si quieres ejes centrados
# ax.axhline(0, color='gray', linewidth=0.5)
# ax.axvline(0, color='gray', linewidth=0.5)

# Mostrar leyenda si hay más de una serie
ax.legend()

plt.show()
```

# 1) Scatter plot
- Datos: dos arrays de igual tamaño
- Ejemplo: x = [1, 2, 3], y = [4, 5, 6]
- Uso: relación entre dos variables numéricas

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = x + np.random.normal(0, 0.5, size=x.shape)

fig, ax = plt.subplots()
ax.scatter(x, y)
plt.show()
```

# 2) Line plot
- Datos: dos arrays de igual tamaño
- Ejemplo: x = [1, 2, 3], y = [4, 5, 6]
- Uso: evolución o tendencia de las dos vartiables

```python
x = np.linspace(0, 10, 100)
y = x + np.random.normal(0, 0.5, size=x.shape)

fig, ax = plt.subplots()
ax.plot(x, y)
```

# 3) Bar plot
- Datos: etiquetas (categorias) y valores
- Ejemplo: etiquetas = ['A', 'B', 'C'], valores = [4, 5, 6]
- Uso: comparar cantidades enrtre las categorias

```python
etiquetas = ["FER","RODRIGO","JUAN"]
valores = [1,2,3]
fig,ax = plt.subplots()
ax.bar(etiquetas,valores)
```

# 4) Histograma
- Datos: un array de valores continuos
- Ejemplo: valores = [1, 2, 2, 3, 4, 5]
- Uso: distribución de frecuencias

```python
data = np.random.randn(1000)

fig,ax = plt.subplots()
ax.hist(data )
plt.show()
```

# 5) Caja de bigotes
datos: lista de arrays (grupos de datos)
Ejemplo: datos = [[1, 2, 3], [4, 5, 6]]
- Uso : resumen estadístico (mendiana y cuartiles)

```python
data = [np.random.randn(1000), np.random.randn(1000)]

fig,ax = plt.subplots()
ax.boxplot(data)
plt.show()
```

# 6) Violín
- Datos: similar a caja de bigotes, pero con densidad
- Ejemplo: datos = [np.random.randn(1000), np.random.randn(1000)]
- Uso: visualizar la distribución de los datos

```python
import seaborn as sns


data = [np.random.randn(1000), np.random.randn(1000)]

sns.violinplot(data=data)
plt.show()
```

# 7) Gráfica de pastel
- Datos: etiquetas y valores
- Ejemplo: etiquetas = ['A', 'B', 'C'], valores = [4, 5, 6]
- Uso: proporciones entre categorías

```python
etiquetas = ["FER","RODRIGO","JUAN"]
valores = [3,2,1]

fig,ax = plt.subplots()
ax.pie(valores, labels=etiquetas)
plt.show()
```

# 8) Quiver o campo de vectores
- Datos: coordenadas de inicio y fin de los vectores (x, y, u, v)
- Ejemplo: x = [0, 1], y = [0, 1], u = [1, 0], v = [0, 1]
- Uso: visualizar campos vectoriales

```python
X,Y = np.meshgrid(np.linspace(-3, 3, 10), np.linspace(-3, 3, 10))
U = -Y
V = X

fig,ax = plt.subplots()
ax.quiver(X,Y,U,V, color='blue',)
plt.show()
```

# 9) Streamplot
- Datos: coordenadas de inicio y fin de los vectores (x, y, u, v)
- Ejemplo: x = [0, 1], y = [0, 1], u = [1, 0], v = [0, 1]
- Uso: visualizar campos vectoriales de manera continua

```python
X,Y = np.meshgrid(np.linspace(-3, 3, 10), np.linspace(-3, 3, 10))
U = -Y
V = X


fig,ax = plt.subplots()
ax.streamplot(X, Y, U, V)
plt.show()
```
