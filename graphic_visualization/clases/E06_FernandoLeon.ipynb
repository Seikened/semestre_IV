```python
import matplotlib.pyplot as plt
import numpy as np

# version de matplotlib
import matplotlib
print(matplotlib.__version__)
```

```python
# ==================================================
# Modo funcional
# ==================================================
x = np.linspace(1,10,100)
y = np.sin(x)
y2 = np.cos(x)
```

```python
# Recibe una lista de valores en x y en y
plt.plot(x, y)
plt.plot(x, y2)
plt
#plt.show()
plt.scatter(x, y, color="lightgreen", label="sin(x)")
# Abajo
plt.xlabel("x")
plt.ylabel("sin(x)", rotation=0, labelpad=5)
plt.title("Ejemplo 1 de grafica", fontsize=20, color="red")
plt.grid(axis="y")
plt.xticks(np.arange(0, 11, 1))
plt.yticks(np.arange(-1, 1.1, 0.1))
plt.legend(loc="lower left")
plt.show()
```

```python
# ==================================================
# Modo orientado a objetos opcion 1
# ==================================================
fig = plt.figure(figsize=(8,4), layout="constrained")


ax1 = fig.add_subplot(121)
ax1.annotate('local max', xy=(np.pi/2, 1), xytext=(2, 1.2),
             arrowprops=dict(facecolor='black', arrowstyle="->"))

ax2 = fig.add_subplot(122)

# ax1
ax1.plot(x, y, color="#FF5733", linestyle="--",marker="+",markersize=7)
ax1.set_title("sin(x)")
ax1.set_xlabel("x")
ax1.set_ylabel("sin(x)")

# ax2
ax2.scatter(x, y2, color="lightgreen")
ax2.grid(axis="y",linewidth=2, alpha=0.2, linestyle="--")
ax2.set_title("cos(x)")
ax2.set_xlabel("x")
ax2.set_ylabel("cos(x)")

plt.tight_layout()
plt.show()
```

```python
# ==================================================
# Modo orientado a objetos opcion 2
# ==================================================

fig, axes = plt.subplots(nrows=2,ncols=2, figsize=(12,12))
fig.suptitle("Titulo principal")

plt.style.use("seaborn-v0_8-darkgrid")
# Eje 1
axes[0][0].plot(x, y)
axes[0][0].set_title("sin(x)")
axes[0][0].set_xlabel("")
axes[0][0].set_ylabel("")
axes[0][0].grid(axis="y")


# Eje 2
axes[0][1].scatter(x, y2, color="lightgreen")
axes[0][1].set_title("cos(x)")
axes[0][1].set_xlabel("")
axes[0][1].set_ylabel("")
axes[0][1].grid(axis="x")



axes[1][0].hist([1,2,2,3,3,3,4,5], bins=5, color="lightblue")
axes[1][0].set_title("Histograma")
axes[1][0].set_xlabel("x")
axes[1][0].set_ylabel("Frecuencia")
axes[1][0].grid(axis="y")
axes[1][0].set_xticks(np.arange(1, 6, 1))



# mostrar
plt.show()

```
