# LÃ­neas de corriente
- Streamlines
Son curvas tangentes al vector del campo en cada punto. Es decir, representan trayectorias

Sea un campo vectorial bidimensional

```python
import numpy as np
import matplotlib.pyplot as plt

x_list = [0, 1, 0, 1, -1, 0, 1, -1]
y_list = [0, 0, 1, 1, 0, -1, -1, 1]

limit = 3

xrango = np.linspace(-limit,limit,10)
yrango = np.linspace(-limit,limit,10)

xx,yy = np.meshgrid(xrango,yrango)

giro = lambda x,y: (-y, x)

g_vector = lambda x,y: (2*x,-2*y)


u_list,v_list = giro(xx,yy)
z_list,w_list = g_vector(xx,yy)
for x,y in zip(x_list, y_list):
    print("\n")
    print(f"Giro de ({x},{y}) = {giro(x,y)}")
    print(f"Vector de ({x},{y}) = {g_vector(x,y)}")


    


fig1, ax1 = plt.subplots(figsize=(5,5))
ax1.quiver(xx, yy, u_list, v_list, angles='xy', scale_units='xy', color='blue')
ax1.set_title("Vectores con Giro")
ax1.set_aspect('equal')
ax1.set_xlim(-5, 5)
ax1.set_ylim(-5, 5)
ax1.grid()


fig2, ax2 = plt.subplots(figsize=(5,5))

ax2.quiver(xx, yy, z_list, w_list, angles='xy', scale_units='xy', color='red')
ax2.set_title("Vectores Transformados")
ax2.set_aspect('equal')
ax2.set_xlim(-limit, limit)
ax2.set_ylim(-limit, limit)
ax2.grid()

plt.show()
```

# Â¿QuÃ© diferencia observas entre los dos campos? Â¿Los vectores giran, se alejan o se acercan al origen?
En la primera figura esta solo girando y en la segunda se estan alejando del eje en y cuando vale 0 pero es raro por que parece que hay una asintota
