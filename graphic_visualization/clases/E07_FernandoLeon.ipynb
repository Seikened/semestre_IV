```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# version de matplotlib
import matplotlib
print(matplotlib.__version__)
```

```python
ruta = '/Users/ferleon/Documents/GitHub/semestre_IV/graphic_visualization/clases/datos/'
```

```python
df = pd.read_csv(ruta +'Advertising.csv')
df.head()
```

```python
# 1 Crear grráfica de líneas Index vs . presupuesto TV | Radio | Perfiodico (Miles de USD)
# Evolución del presupuesto

fix,ax = plt.subplots(figsize=(10,5))



ax.plot(df.index, df['TV'], label='TV', color='brown')
ax.plot(df.index, df['Radio'], label='Radio', color='pink')
ax.plot(df.index, df['Newspaper'], label='Newspaper', color='blue')
ax.set_title('Evolución del presupuesto')
ax.set_xlabel('ID de la campaña')
ax.set_ylabel('Presupuesto (Miles de USD)')
ax.grid(axis='y', linestyle='--', alpha=0.7)
ax.legend()
plt.show()
```

```python
# 2 Crear gráfica de dispersión Presupuesto (TV | Radio | Periódico) vs Ventas
# Presupuesto vs Ventas
fig,ax = plt.subplots(figsize=(10,5), ncols=3, nrows=1, layout='constrained')


fig.suptitle("Presupuesto vs Ventas", fontsize=16)
ax[0].scatter(df['TV'],df['Sales'], color='brown')
ax[0].set_title('TV vs Ventas')
ax[0].set_xlabel('Presupuesto TV')
ax[0].set_ylabel('Ventas')



ax[1].scatter(df['Radio'],df['Sales'], color='pink')
ax[1].set_title('Radio vs Ventas')
ax[1].set_xlabel('Presupuesto Radio')
ax[1].set_ylabel('Ventas')


ax[2].scatter(df['Newspaper'],df['Sales'], color='blue')
ax[2].set_title('Newspaper vs Ventas')
ax[2].set_xlabel('Presupuesto Newspaper')
ax[2].set_ylabel('Ventas')

plt.legend()
plt.show()
```

```python
# 3 Crear gráfica de barras Presupuesto promedio TV | Radio | Periódico
# Comparación presupuesto  en diferentes medios
fig,ax = plt.subplots(figsize=(10,5), layout='constrained')
fig.suptitle("Comparación presupuesto  en diferentes medios", fontsize=16)




columnas = ['TV', 'Radio', 'Newspaper']
colores = ['brown', 'pink', 'blue']
promedios = df[columnas].mean()
ax.bar(columnas,promedios, color=colores, alpha=0.7)
ax.set_title('Presupuesto promedio')
ax.set_xlabel('Medios')
ax.set_ylabel('Presupuesto (Miles de USD)')
ax.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

```python
# 4 Crear historgrama de ventas
# Observar la distribución de la variable ventas

fig ,ax = plt.subplots(figsize=(10,5), layout='constrained')
fig.suptitle("Distribución de la variable ventas", fontsize=16)


ventas = df['Sales']
ventas.agg(['min','max'])
intervalos = np.arange(0, 30, 2)

ax.hist(df['Sales'], color='purple',bins=intervalos, alpha=0.6, edgecolor='black', linewidth=0.2)
ticks = (intervalos[:-1] + intervalos[1:]) / 2
etiquetas = [f'[{intervalos[i]}, {intervalos[i+1]})' for i in range(len(intervalos)-1)]
ax.set_xticks(ticks)
ax.set_xticklabels(etiquetas)

ax.set_title('Histograma de ventas')
ax.set_xlabel('Ventas')
ax.set_ylabel('Frecuencia')




plt.legend()
plt.show()
```

```python
# GRafico de  3D para visualizar la relación entre:
# TV + Radio vs Sales, TV + Newspaper vs Sales, Radio + Newspaper vs Sales
columnas = [df['TV'], df['Radio'], df['Sales']]
fig, ax = plt.subplots(figsize=(20, 15), layout='constrained', subplot_kw={'projection': '3d'}, ncols=1, nrows=3)
fig.suptitle("Relación entre TV + Radio vs Sales, TV + Newspaper vs Sales, Radio + Newspaper vs Sales", fontsize=16)


ax[0].scatter(df['TV'], df['Radio'], df['Sales'], color='brown')
ax[0].set_title('TV + Radio vs Sales')
ax[0].set_xlabel('TV')
ax[0].set_ylabel('Radio')
ax[0].set_zlabel('Sales')

ax[1].scatter(df['TV'], df['Newspaper'], df['Sales'], color='pink')
ax[1].set_title('TV + Newspaper vs Sales')
ax[1].set_xlabel('TV')
ax[1].set_ylabel('Newspaper')
ax[1].set_zlabel('Sales')



ax[2].scatter(df['TV'], df['Newspaper'], df['Sales'], color='blue')
ax[2].set_title('Radio + Newspaper vs Sales')
ax[2].set_xlabel('Radio')
ax[2].set_ylabel('Newspaper')
ax[2].set_zlabel('Sales')
plt.legend()
plt.show()

```
