```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python

matriz = np.array([[2,5,8],
                   [-1,4,-7],
                   [0,-3,6]])
fig, ax = plt.subplots()
s = ax.imshow(matriz, cmap="bwr")
fig.colorbar(s, ax=ax, label='Valor escala')

```

```python
x1 = np.random.normal(0, 1, 50)
x2 = np.random.normal(0, 1, 500)
x3 = np.random.normal(0, 1, 5000)

y1 = np.random.normal(0, 1, 50)
y2 = np.random.normal(0, 1, 500)
y3 = np.random.normal(0, 1, 5000)

fig,ax = plt.subplots(figsize=(10,10), ncols=3, nrows=3, layout='constrained')

fig.suptitle("Presupuesto vs Ventas", fontsize=16)

ax[0, 0].scatter(x1, y1)
ax[0, 1].scatter(x2, y2)
ax[0, 2].scatter(x3, y3)


ax[1, 0].hist2d(x1, y1, bins=20)
ax[1, 1].hist2d(x2, y2, bins=20)
ax[1, 2].hist2d(x3, y3, bins=20)


ax[2,0].hexbin(x1, y1, gridsize=20)
ax[2,1].hexbin(x2, y2, gridsize=20)
ax[2,2].hexbin(x3, y3, gridsize=20)
```
